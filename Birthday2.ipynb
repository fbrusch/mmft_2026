```python
def frequency(birthday, people):
    return sum([x == birthday for x in people])
```

```python
def check_triples(people):
    return any([frequency(x, people) >= 3 for x in range(365)])
```

```python
sum([check_triples(generate_class()) for x in range(100000)])/100000
```

```python
def check_couples(people):
    return any([frequency(x, people) >= 2 for x in range(365)])
```

```python
def check_n_collisions(n, people):
    return any([frequency(x, people) >= n for x in range(365)])
    
```

```python
def generate_class(n):
    return [randint(0, 364) for x in range(n)]
```

```python
def estimate_couples_probability(n, n_of_experiments):
    return sum([check_n_collisions(2, generate_class(n)) for x in range(n_of_experiments)])/n_of_experiments
```
