### Monte Carlo Estimate

Now we repeat the experiment many times. Each trial checks whether at least two people share a birthday. Averaging many trials approximates the probability.

Docs reference:
- Boolean operations and comparisons: https://docs.python.org/3/library/stdtypes.html#boolean-operations-and-or-not
- `sum()` over booleans: https://docs.python.org/3/library/functions.html#sum

```python
[len(set([floor(random()*365) for x in range(25)])) < 25 for x in range(100)]
```

```python
sum([len(set([floor(random()*365) for x in range(26)])) < 26 for x in range(10000)])/10000
```
